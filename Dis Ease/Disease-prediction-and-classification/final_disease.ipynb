### Custom Gaussian Naive Bayes

```python
XX=  df.drop(columns =['diseases'])
YY = df['diseases']
XX_train, XX_test, YY_train, YY_test = train_test_split(XX, YY, test_size=0.2, random_state=44)
```

```python
import numpy as np

class CustomGaussianNB:
    def __init__(self):
        self.classes = None
        self.mean = {}
        self.var = {}
        self.priors = {}

    def fit(self, X, y):
        self.classes = np.unique(y)
        for cls in self.classes:
            X_cls = X[y == cls]
            self.mean[cls] = X_cls.mean(axis=0)
            self.var[cls] = X_cls.var(axis=0)
            self.priors[cls] = len(X_cls) / len(X)

    def cal_log_likelihood_vectorized(self, mean, var, X):
        eps = 1e-9
        coeff = -0.5 * np.log(2 * np.pi * var + eps)
        exponent = -((X - mean) ** 2) / (2 * var + eps)
        return coeff + exponent

    def cal_log_posterior_vectorized(self, X):
        log_posteriors = []
        for cls in self.classes:
            prior = np.log(self.priors[cls] + 1e-9)
            log_likelihoods = np.sum(
                self.cal_log_likelihood_vectorized(self.mean[cls], self.var[cls], X), axis=1
            )
            log_posteriors.append(prior + log_likelihoods)
        return np.column_stack(log_posteriors)

    def predict(self, X):
        log_posteriors = self.cal_log_posterior_vectorized(X)
        return self.classes[np.argmax(log_posteriors, axis=1)]

    def predict_proba(self, X):
        log_posteriors = self.cal_log_posterior_vectorized(X)
        exp_posteriors = np.exp(log_posteriors - np.max(log_posteriors, axis=1, keepdims=True))
        probabilities = exp_posteriors / np.sum(exp_posteriors, axis=1, keepdims=True)
        return probabilities
```

```python
model2 = CustomGaussianNB()
model2.fit(XX_train.to_numpy(), YY_train.to_numpy())
```

```python
def predict_disease(user_symptoms):
    feature_vector = np.zeros(len(XX.columns))

    for symptom in user_symptoms:
        if symptom in XX.columns:
            feature_vector[XX.columns.get_loc(symptom)] = 1

    probabilities = model2.predict_proba([feature_vector])[0]
    top_3_indices = probabilities.argsort()[-3:][::-1]
    top_3_diseases = [(model2.classes[i], probabilities[i]) for i in top_3_indices]

    return top_3_diseases
```

```python
user_symptoms = [
'lip sore', 'vulvar sore', 
    'hip stiffness or tightness', 'mouth pain', 'arm weakness'
]
```

```python
predicted_diseases = predict_disease(user_symptoms)
print("Top 3 Predicted Diseases:")
for disease, probability in predicted_diseases:
    print(f"{disease}: {probability * 100:.2f}%")
```

```python
predicted_labels = model2.predict(XX_test.to_numpy())
```

```python
accuracy = accuracy_score(YY_test, predicted_labels)
print(f"Accuracy of Custom Naive Bayes: {accuracy * 100:.2f}%")
```

### Custom Decision Tree Classifier

```python
df_sampled = df.sample(frac=0.1, random_state=42)
x = df_sampled.drop(columns=['diseases'])
y = df_sampled['diseases']
```

```python
x = x.astype(np.float32)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
class CustomDecisionTreeClassifier:
    def __init__(self, max_depth=None, min_samples_split=2):
        """Initialize the custom Decision Tree Classifier."""
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.tree = None
        self.classes_ = None

    def fit(self, X, y):
        X = np.array(X) 
        y = np.array(y)  
        self.classes_ = np.unique(y)
        self.tree = self._build_tree(X, y, depth=0)

    def _build_tree(self, X, y, depth):
        """Recursively build the decision tree."""
        num_samples, num_features = X.shape

        if len(set(y)) == 1 or num_samples < self.min_samples_split or (self.max_depth and depth >= self.max_depth):
            return max(set(y), key=list(y).count)

        best_feature, best_threshold = self._best_split(X, y)

        if best_feature is None:
            return max(set(y), key=list(y).count)

        left_indices = X[:, best_feature] <= best_threshold
        right_indices = X[:, best_feature] > best_threshold

        left_subtree = self._build_tree(X[left_indices], y[left_indices], depth + 1)
        right_subtree = self._build_tree(X[right_indices], y[right_indices], depth + 1)

        return {"feature": best_feature, "threshold": best_threshold, "left": left_subtree, "right": right_subtree}

    def _best_split(self, X, y):
        num_samples, num_features = X.shape
        best_gain, best_feature, best_threshold = 0, None, None

        for feature in range(num_features):
            thresholds = np.unique(X[:, feature])
            for threshold in thresholds:
                gain = self._information_gain(X[:, feature], y, threshold)
                if gain > best_gain:
                    best_gain, best_feature, best_threshold = gain, feature, threshold

        return best_feature, best_threshold if best_gain > 0 else None

    def _entropy(self, y):
        """Calculate entropy of labels."""
        _, counts = np.unique(y, return_counts=True)
        probabilities = counts / counts.sum()
        return -np.sum(probabilities * np.log2(probabilities + 1e-9)) 

    def _information_gain(self, feature_column, y, threshold):
        left_indices = feature_column <= threshold
        right_indices = feature_column > threshold

        if sum(left_indices) == 0 or sum(right_indices) == 0:
            return 0

        entropy_before = self._entropy(y)
        left_entropy = self._entropy(y[left_indices])
        right_entropy = self._entropy(y[right_indices])

        left_weight = sum(left_indices) / len(y)
        right_weight = sum(right_indices) / len(y)

        entropy_after = left_weight * left_entropy + right_weight * right_entropy

        return entropy_before - entropy_after

    def predict(self, X):
        X = np.array(X)  
        return [self._traverse_tree(self.tree, sample) for sample in X]

    def _traverse_tree(self, tree, sample):
        if not isinstance(tree, dict):
            return tree 

        feature = tree["feature"]
        threshold = tree["threshold"]

        if feature >= len(sample): 
            return tree

        if sample[feature] <= threshold:
            return self._traverse_tree(tree["left"], sample)
        else:
            return self._traverse_tree(tree["right"], sample)

    def predict_proba(self, X):
        predictions = self.predict(X)
        proba = np.zeros((len(predictions), len(self.classes_)))

        for i, pred in enumerate(predictions):
            class_index = np.where(self.classes_ == pred)[0][0]
            proba[i, class_index] = 1 
        return proba

    def predict_top_n(self, X, n=3):
        proba = self.predict_proba(X)
        top_n_predictions = []

        for row in proba:
            top_indices = row.argsort()[-n:][::-1]
            top_n_predictions.append([(self.classes_[i], row[i]) for i in top_indices])

        return top_n_predictions
```

```python
model = CustomDecisionTreeClassifier(max_depth=None, min_samples_split=5)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
def predict_disease2(user_symptoms):
    feature_vector = np.zeros(len(symptoms_list))

    for symptom in user_symptoms:
        if symptom in symptoms_list:
            feature_vector[symptoms_list.index(symptom)] = 1

    probabilities = model.predict_proba([feature_vector])[0]

    top_1_indices = probabilities.argsort()[-1:][::-1]
    top_1_diseases = [(model.classes_[i], probabilities[i]) for i in top_1_indices]

    return top_1_diseases
```
